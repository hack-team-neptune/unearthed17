### make all plots in the dictionary

```python
line_styles=['-', '--', ':', ';']
colours=[]
for name, hex in mp.colors.cnames.iteritems():
    colours.append(name)
```

```python
for i, y_vals in enumerate(data_cols):
    
    fig = plt.Figure(figsize=(4,4))
    
    
    plt.plot(np.arange(len(data.index))*5/60/24, data[data_cols[y_vals]].values, label=y_vals, ls=line_styles[0], 
            c=colours[i])
    

    
    plt.xlabel('Date')
    plt.ylabel('Value')
    
    plt.legend()
    
    plt.tight_layout()
    
    plt.savefig('figures/corr_'+str(y_vals)+'.png', dpi=400)
    
    plt.close()
    
    
    
    
```
